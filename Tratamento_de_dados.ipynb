```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Carregando o dataset
automobile = pd.read_csv("/content/Meu_Drive/MyDrive/Colab Notebooks/Automobile_data.csv", sep=',')
```

```python
automobile.head(20)
```

```python
# Como não é possível determinar quais são os valores faltantes na coluna normalized-losses, a mesma será excluída
# Para os fins da nosso objetivo, a coluna normalized-losses não trará significância e manter-la pode causar erro de cálculos futuros.
automobile_new = automobile.drop("normalized-losses", axis="columns" )
```

```python
automobile_new.head(20)
```

```python
list(automobile_new.columns)
```

```python
#Transformando todos os campos com "?" por nan
automobile_new = automobile_new.replace('?', np.nan)
```

```python
#Somando todos os nan de todas as colunas
#Esse metódo facilita o processo de tratamento de dados, visto que sabemos onde é necessário realizar alterações
automobile_new.isna().sum()
```

```python
# Aqui foi criado uma tabela alternativa com os valores númericos da coluna número de portas, para que seja possível alterar os valores nulos
automobile_alter = automobile_new.replace("two", 2)
automobile_alter = automobile_alter.replace("four", 4)
```

```python
# A mediana encontrada irá substituir os valores NaN
automobile_alter['num-of-doors'].median()
```

```python
automobile_new['num-of-doors'] = automobile_new['num-of-doors'].replace(np.nan,4)
```

```python
automobile_new['bore'].dtypes
```

```python
columns_to_convert=['bore', 'stroke','horsepower','peak-rpm','price']
```

```python
#utilizando um laço for para converter todas as colunas para numericos e exibir valores não numericos
# A conversão foi feita para substituir os valores NaN pela média geral
for i in columns_to_convert:
  automobile_new[i]=pd.to_numeric(automobile_new[i], errors='coerce')
```

```python
#Exemplo para mostrar que os valores foram trocados
automobile_new['bore'].dtypes
```

```python
#Substituindo os NaN pela média da coluna
# O fillna funciona para substituir valores ausentes por algo
for i in columns_to_convert:
   mean_colunm = automobile_new[i].mean()
   automobile_new[i]=automobile_new[i].fillna(mean_colunm)
```

```python
#Verificando novamente se há NaN
#Como percebe-se, não há.
automobile_new.isna().sum()
```

```python
#Salvando o dataset novo para fazer a análise exploratória dos dados
automobile_new.to_csv('automobile_new.csv', index=False)
```
